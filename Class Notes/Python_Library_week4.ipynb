```python
#create a 5x5 matrix with values 1,2,3,4 just below the diagonal

v = np.zeros((5,5))
print(v)
print()
values = [1, 2, 3, 4]
for i in range(1, 5):
    v[i, i - 1] = values[i - 1]
print(v)
```
