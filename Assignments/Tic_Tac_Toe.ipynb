```python
import random

def print_board(board):
    for r in board:
      print(" | ".join(r) )
      print("---------")

def check_winner(board, player):
    for i in range(3):
        if all(board[i][j] == player for j in range(3)):
            return True
        if all(board[j][i] == player for j in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    else:
      return False

def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)

def get_empty_cells(board):
    empty_cells = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                empty_cells.append((i, j))
    return empty_cells

def computer_input(board):
    empty_cells = get_empty_cells(board)
    return random.choice(empty_cells)

def play_game():
  board = [[' ' for _ in range(3)] for _ in range(3)]
  user_name = input('Enter your name: ')
  user_symbol = input('Choose your symbol X or O: ').upper()
  if user_symbol == 'X':
    computer_symbol = 'O'
  else:
    computer_symbol = 'X'
  print("Welcome to Tic Tac Toe Game", user_name,'!')
  print('The symbol for computer is: ', computer_symbol)

  print_board(board)

  while True:
    row = int(input("Enter row (1-3): "))-1
    col = int(input("Enter column (1-3): "))-1

    if board[row][col] == ' ':
      board[row][col] = user_symbol
    else:
      print("Cell not empty. Try again.")
      continue

    print_board(board)

    if check_winner(board, user_symbol):
      print("Congratulations, You won!")
      break
    elif is_board_full(board):
      print("It's a draw!")
      break

    computer_row, computer_col = computer_input(board)
    board[computer_row][computer_col] = computer_symbol

    print("Computer's move:")
    print_board(board)

    if check_winner(board, computer_symbol):
      print("Sorry, Computer won! Try again")
      break
    elif is_board_full(board):
      print("It's a draw!")
      break

def play_again():
  user_input = True

  while user_input:
    play_again = input("Do you want to play again? (Y/N): ").upper()
    if play_again == "Y":
      play_game()
    else:
      print("Thank you for playing!")
      break

play_game()
play_again()
```
